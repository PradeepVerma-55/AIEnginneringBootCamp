```
A command-line tool that:
```

---

## 💡 Real-World Analogy

> Market Cap = **Total value of a company**  
> Formula: `Stock Price × Total Shares = Market Cap`  
> e.g. Apple at $180 × 15 billion shares = $2.7 Trillion

```python
class Company:
    def __init__(self, name: str, stock_price: float, shares: int):
        self.name        = name
        self.stock_price = stock_price
        self.shares      = shares

    def market_cap(self) -> float:
        return self.stock_price * self.shares

    def display(self):
        cap = self.market_cap()
        cap_in_trillions = cap / 1_000_000_000_000
        print(f"{self.name:<12} | ${self.stock_price:<8} | ${cap_in_trillions:.2f}T")
```

## 🧮 Step 3 — Build the Calculator

```python
# market_cap_calculator.py

class Company:
    def __init__(self, name: str, stock_price: float, shares: int):
        self.name        = name
        self.stock_price = stock_price
        self.shares      = shares

    def market_cap(self) -> float:
        return self.stock_price * self.shares

    def display(self):
        cap_in_trillions = self.market_cap() / 1_000_000_000_000
        print(f"{self.name:<12} | ${self.stock_price:<8} | ${cap_in_trillions:.2f}T")


class MarketCapCalculator:
    def __init__(self):
        self.companies = []          # empty list to start

    def add_company(self, company: Company):
        self.companies.append(company)

    def rank_by_market_cap(self):
        # sort largest → smallest
        return sorted(
            self.companies,
            key=lambda c: c.market_cap(),
            reverse=True
        )

    def display_all(self):
        print("=" * 45)
        print(f"{'Company':<12} | {'Price':<9} | Market Cap")
        print("=" * 45)
        ranked = self.rank_by_market_cap()
        for i, company in enumerate(ranked, start=1):
            print(f"#{i} ", end="")
            company.display()
        print("=" * 45)
```

```csharp
// C# equivalent
public class MarketCapCalculator
{
    private List<Company> _companies = new();

    public void AddCompany(Company company)
        => _companies.Add(company);

    public List<Company> RankByMarketCap()
        => _companies.OrderByDescending(c => c.MarketCap()).ToList();

    public void DisplayAll()
    {
        Console.WriteLine(new string('=', 45));
        Console.WriteLine($"{"Company",-12} | {"Price",-9} | Market Cap");
        Console.WriteLine(new string('=', 45));

        var ranked = RankByMarketCap();
        for (int i = 0; i < ranked.Count; i++)
        {
            Console.Write($"#{i + 1} ");
            ranked[i].Display();
        }
        Console.WriteLine(new string('=', 45));
    }
}
```

````python
class Company:
    def __init__(self,name,stock_price,no_of_shares):
        self.name= name
        self.stock_price= stock_price
        self.no_of_shares= no_of_shares

    def market_cap(self):
        return self.stock_price * self.no_of_shares

    def display_info(self):
        market_cap = self.market_cap()  
        print(f"Company Name: {self.name}")
        print(f"Stock Price: ${self.stock_price}")  
        print(f"Number of Shares: {self.no_of_shares}")
        print(f"Market Capitalization: ${market_cap}")

# Example usage

apple = Company("Apple Inc.", 150, 1000000)
market_cap =apple.market_cap()
print(f"Market Capitalization of {apple.name}: ${market_cap}")
apple.display_info()
````

# Market Cap Calculator

````python
class MarketCapCalulator:
    def __init__(self):
        self.companies = []

    def add_company(self,company):
        self.companies.append(company)

    def display_all_companies(self):
        for company in self.companies:
            company.display_info()

    def calculate_total_market_cap(self):
        total_market_cap = 0
        for company in self.companies:
            total_market_cap += company.market_cap()
        return total_market_cap

    def display_market_caps(self):
        for company in self.companies:
            market_cap = company.market_cap()
            print(f"{company.name} Market Cap: ${market_cap}")

# Example usage
market_cap_calculator = MarketCapCalulator()
apple = Company("Apple Inc.", 150, 1000000)
microsoft = Company("Microsoft Corporation", 250, 500000)
amazon = Company("Amazon.com, Inc.", 3500, 200000)
alphabet = Company("Alphabet Inc.", 2800, 300000)
facebook = Company("Facebook, Inc.", 350, 400000)
tesla = Company("Tesla, Inc.", 700, 600000)
nvidia = Company("NVIDIA Corporation", 220, 800000)
jpmorgan = Company("JPMorgan Chase & Co.", 160, 700000)
johnson = Company("Johnson & Johnson", 170, 900000)
visa = Company("Visa Inc.", 230, 1000000)
market_cap_calculator.add_company(apple)
market_cap_calculator.add_company(microsoft)
market_cap_calculator.add_company(amazon)
market_cap_calculator.add_company(alphabet)
market_cap_calculator.add_company(facebook)
market_cap_calculator.add_company(tesla)
market_cap_calculator.add_company(nvidia)
market_cap_calculator.add_company(jpmorgan)
market_cap_calculator.add_company(johnson)
market_cap_calculator.add_company(visa)
market_cap_calculator.display_all_companies()
total_market_cap = market_cap_calculator.calculate_total_market_cap()
print(f"Total Market Capitalization: ${total_market_cap}")
    
````
