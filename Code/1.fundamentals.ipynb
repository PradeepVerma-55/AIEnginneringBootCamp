#  Functions


````python
# Python functions

def generate_response(prompt, model="gpt-4")->str :
    response = f"Response generated from model {model} for prompt: {prompt}"
    return response

# Example usage
prompt = "What is the capital of France?"
response = generate_response(prompt)
print(response)

````

# Classes & OOP


````python
class Dog:
    def __init__(self,name):
        self.name = name
    
    def bark(self):
        return f"My Dog name is :{self.name}"


# Example usage
my_dog = Dog("Buddy")
print(my_dog.bark())

````

````python
# Lets add one more property to the Dog class

class Dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def bark(self):
        return f"My Dog name is : {self.name} and age is : {self.age} years"


# Example usage
my_dog = Dog("Buddy", 3)
print(my_dog.bark())
````

# Lets Build LLM Client...

````python
class LLMClient: 
    def __init__(self,model:str,key:str):
        self.model = model
        self.key = key

    def chat(self,prompt:str)->str:
         return f"{self.model} Response to: {prompt}"

# Example usage
client = LLMClient(model="gpt-4", key="my-secret-key-12334454")
response = client.chat("What is the capital of France?")
print(response)
````
